## Material and method

### Study area
The Adolpho Ducke Forest Reserve (02o 55 'S and 59o 59' W) is located north of Manaus, at km 23 of AM-010 and covers an area greater than 100 km2 (Figure S1). The area is administrated by the National Institute of Amazonian Research (INPA), and is predominantly covered with old-growth forest with a closed canopy.  Tree total height varies between 30 and 47 m (Ribeiro et al. 1999) reaching up to 60 m, as indicated by LiDAR data. Forest structure and soil type vary with local undulating topography (Castilho et al. 2006). Terrain elevation varies from 40 m to 120 m across SRTM - NASA's Shuttle Radar Topography Mission (See Remote Sensing section, USGS, 2004) and from 55 to 105 across LiDAR data.

## Data 
### Biomass data
The allometric dataset includes 131 trees > 5 cm DBH harvested in an area ~50 km northwest of the study area (Silva 2007, Marra et al., 2016).  Total above ground fresh mass was measured for each harvested tree.  For 31 trees, wood samples were collected to calculate the water content of the stem, coarse branches > 10 cm diameter, fine branches < 10 cm diameter, leaves, flowers and fruits. These data were used to calculate the dry mass content of each tree and the respective uncertainty. The variation and mean carbon content (CCdry) of dry tissues were obtained from Silva (2007). 


### Forest inventory
In April 2013, twenty-six forest inventory plots of 40 m x 125 m were installed in an area of 3 km x 4 km (1,247 ha) that was previously overflown by an airborne LiDAR (Light Detection and Ranging) system in the previous year (See Remote Sensing section, Figure 2). Plots were installed by laying out a 125 m tape to make a central trail for each plot, and censuring trees within 20 m on both sides, using a laser rangefinder (Laser Technology TruPulse 360B) to determine distance to the tape.  We recorded all live and dead trees (including palms) with diameter at breast height (DBH) >10 cm. The DBH of live trees was measured with a diameter tape (1 cm precision). The geolocations were taken for all trees in each plot using a Global Navigation Satellite System (GNSS) Garmin 62s. For trees with tall buttresses, trunk diameters were estimated visually by expert field workers. We plotted a locally referenced field map for each plot in Cartesian coordinates, with tree locations described by their position on the 20 m (X) and 125 m (Y) side of the plot. Distances were measured using a laser rangefinder device (Laser Technology TruPulse 360B.).  The spatial variation of GNSS coordinates of each tree were used to calculate the rotation and translation required to transform from local Cartesian coordinates to global geographic coordinates. 
We evaluated plot-georeferencing associated errors by comparing the resulting geographic with the high-precision GNSS coordinates collected at 19 points using a topography GNSS (Techgeo Zenite). These points were collected at the start and/or end of the central trail of selected plots. We attempted collection at all plots, but problems obtaining a good signal through the canopy from sufficient satellites reduced the number of suitable data points. 

### Remote sensing
We used airborne LiDAR data for an area of 3 km x 4 km (1,247 ha)  acquired within the Embrapa's Sustainable Landscapes Project and are available at: https://www.paisagenslidar.cnptia.embrapa.br/webgis/ (Figure S1). Imagery data were acquired at an altitude of 1,500 m. Sweeping angle perpendicular to the terrain was used at the time of the flights, and obeyed the maximum of 20º. The covering of the flight lines was 30% (overlap) and showed a high-density point cloud with mean of ~23 pts/m2.
The elevation model derived from SRTM carried out in February 2000 (USGS, 2004). SRTM data were used for the landscape level mapping. Four SRTM tiles were merged and clipped by the Ducke reserve boundary. As Federal protected area, Ducke had no sign of deforestation. Therefore, the STRM data account for the topographic variation but is influenced by variations of canopy height. This highlights the differences in structure among forest types within our study area.

## Data analysis

### Workflow

The quantification and propagation of carbon uncertainty along the upscaling were performed in three levels: (i) quantification of individual tree biomass; (ii) combination of plot-based carbon and canopy height map derived from LiDAR to produced a high resolution carbon map; and (iii) combination of high resolution carbon map and elevation data derived from SRTM to produce a landscape carbon map (Figure 1). We first calculated the uncertainty of each step (allometry, carbon content, trunk diameter, georeferencing and remote sensing (red in Figure 1, objective 1). We defined the best approach for upscaling from field plots to LiDAR, combining stem-localized or crown-distributed estimated carbon in plots and fixed rectangular subplots or segments based on similarity in canopy height (purple rectangles in Figure 1, objective 2). We selected the select two approach to calculate the equations and uncertainties of high resolution and landscape carbon maps (purple axis of fits in Figure 1, objective 2). We finally propagated the uncertainties using Monte Carlo Method based on 1000 simulations for producing 1000 carbon maps (dark red in Figure 1, objectives 3 and 4). We calculated the mean and uncertainty of carbon maps by selecting regions from one pixel to landscape.
